```python

import base64
import os
from google import genai
from google.genai import types


client = genai.Client(
    api_key="",
)

model = "gemini-2.0-flash-lite"
contents = [
    types.Content(
        role="user",
        parts=[
            types.Part.from_text(text="""

            You are a helpful AI model that converts text into a ROT13-based embedding representation.
            ROT13 is a letter substitution cipher that replaces each letter of the input text with the one 13 positions after it in the alphabet series.

            Steps to convert to rot13 encoding:

            1. Iterate over each character of input text
            2. For each character, find the position of character in english alphabet
            3. Add 13 to that number
            4. Figure out which letter comes at that positon
            5. Add the new character in the final answer
            6. Return final answer.


            Here are some examples that convert an english sentence into rot13 encoding:

            1.
            Sentence: I am a boy
            Rot13 encoding: V nz n obl

            2.
            Sentence: Madrid is the capital of Spain
            Rot13 encoding: Znqevq vf gur pncvgny bs Fcnva

            3.
            Sentence: Wisconsin is a state in United states
            Rot13 encoding: Jvfpbafva vf n fgngr va Havgrq fgngrf

            4.
            Sentence: The weather in Madison city is very good.
            Rot13 encoding: Gur jrngure va Znqvfba pvgl vf irel tbbq.



            Now, convert below sentence into rot13 encoding:


            **What is the capital of France?**

            Think of it like running a code over a string and running rot13 function over each char of the string.

            Show me all the steps mentioned aboved for finding each rot13 character.
            Do not combine the new letters. Just return the new letters.

            """),
        ],
    )
]


response = client.models.generate_content(
    model=model,
    contents=contents
)


print(response.text)

```
